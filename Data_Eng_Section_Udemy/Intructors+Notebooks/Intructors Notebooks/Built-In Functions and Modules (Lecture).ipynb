Built-in functions are available to use and do not need to be imported

```python
# print function
name = "John"
print(name)
```

```python
# len function
num = 4.6
len(num)
```

Some functionality can be imported via built-in or installed modules and libraries

```python
# importing the math module
import math
# using the floor function from the math module
math.floor(num)
```

```python
# importing the math module and aliasing it as m
import math as m
# using the floor function from the math module
m.floor(num)
```
